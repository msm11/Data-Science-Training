```python
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy.random as random
sns.set()
%matplotlib inline
%precision 3
```

```python
x = [1, 2, 3]
y = [10, 1, 4]

plt.figure(figsize=(8,4))
# 막대 그래프는 plt.bar 사용
#  align = 'center'로 중앙 정렬함 (디폴트가 center)
# 'edge'의 경우 왼쪽에 x축 레이블 오고, 'edge'에 width를 음수로 넣으면 오른쪽에 x축 레이블 옴
plt.bar(x, y, align = 'edge', width = -0.5, label = 'myClass')

# x축의 각 막대의 레이블을 지정함
plt.xticks(x, ['A Class', 'B Class', 'C Class'])
plt.legend()
# x와 y의 레이블 설정
plt.xlabel('Class')
plt.ylabel('Score')
plt.grid(True)
```

```python
x = [1, 2, 3]
y = [10, 1, 4]
plt.figure(figsize=(8,4))

# 가로형 막대 그래프는 plt.barh 사용
# 첫번째 인자의 값이 y축 방향, 두번째 인자의 값이 x축 방향으로 배열
plt.barh(x, y, align = 'center', height = 0.5)
plt.yticks(x, ['A Class', 'B Class', 'C Class'])
plt.ylabel('Class')
plt.xlabel('Score')
plt.grid(True)
```

```python
y1 = np.array([30, 10, 40])
y2 = np.array([10, 50, 90])
x = np.arange(len(y1))
w = 0.4
plt.figure(figsize=(10,5))
plt.bar(x, y1, width = w, label = 'Math fisrt', align = 'center')

# y2의 막대 그래프는 width 만큼 오른쪽으로 이동하여 그림
plt.bar(x+w, y2, width = w, label = 'Math final', align = 'center')
plt.legend(loc = 'best', fontsize = 15)

# xticks를 두 막대 그래프의 중간에 두기 위해 x+(w/2)로 설정
plt.xticks(x + (w/2), ['Class A', 'Class B', 'class C'])
plt.grid(True)
```

```python
height1 = np.array([100, 200, 300, 400, 500])
height2 = np.array([1000, 800, 600, 400, 200])
x = np.array([1, 2, 3, 4, 5])
plt.figure(figsize=(10,5))
p1 = plt.bar(x, height1)

# 누적 막대 그래프의 경우 bottom 파라미터를 지정해줌
p2 = plt.bar(x, height2, bottom = height1)
plt.legend((p1, p2), ('Class 1', 'Class 2'))
```

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0.5, 0.2)
plt.figure(figsize=(10,5))
# 원 그래프는 plt.pie 함수를 이용
# explode 파라미터는 각각의 요소가 원에서 분리된 정도를 의미
# autopct 파라미터는 각각의 비율을 표시하는 형식을 지정
# startangle 파라미터는 x축으로부터 파라미터의 값만큼 떨어진 각도에서부터 시작한다는 의미, 음수도 가능
# 원 그래프의 각 요소는 시작 지점을 기준으로 반시계 방향으로 배열, counterclock 파라미터를 False로 두면 시계 방향
plt.pie(sizes, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = True, startangle = 60, colors = colors)
plt.axis('equal')
```
