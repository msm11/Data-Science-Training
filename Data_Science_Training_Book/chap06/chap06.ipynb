```python
"""
 pd.cut은 처음 argument의 data를 2번째 arugment의 구간으로 나누어
 데이터가 2번째 argument의 구간 중 어디에 속하는지를 나타냄
 또, 처음 argument에는 1 dimensional array가 와야 함
 cut 함수의 경우 '( , ]'의 형태로 값을 리턴하는데, 왼쪽 범위는 포함되지 않고 오른쪽 범위는 포함된다는 의미를 가짐 
 열린 구간, 닫힌 구간의 설정은 cut 함수의 left 옵션과 right 옵션으로 변경 가능
"""
birth_year_bins = [1980, 1985, 1990, 1995, 2000]
birth_year_cut_data = pd.cut(df1['birth_year'], birth_year_bins, right=False)
birth_year_cut_data
```

```python
# 아래와 같은 방식으로도 pd.cut 사용 가능
# 이 경우 전체 구간을 2개로 분할한다는 의미
pd.cut(df1.birth_year, 2)
```

```python
"""
pd.cut('1d array or Series', 'n')은 데이터의 전체 길이를 n등분 하여 나눔
pd.qcut('1d array or Series', 'n')은 데이터를 분위수를 기준으로 n등분 하여 나눔
예를 들어 pd.qcut('1d array or Series', 5)는 데이터를 5개로 분할하는데
5개 구간에 속한 데이터의 수가 동일함 (1~2개의 오차 있을 수 있음)
"""
print(pd.value_counts(pd.cut(df1['birth_year'], 5)).sort_index())
print()
pd.cut(df1['birth_year'].sort_values(), 5)
```

```python
# 특정 열만 결측 데이터를 고려하여 활용하는 방법
df[[0,1]].dropna()
```

```python
# fillna() 메서드는 NaN의 값을 argument의 값으로 모두 대체한
# DataFrame을 리턴
df.fillna(0)
```

```python
# fillna() 메서드에서  method의 argument를 'ffill'로 설정 시
# NaN를 인접 데이터의 값 (이전 행의 데이터 값)으로 대체한 DataFrame을 리턴
df.fillna(method = 'ffill')
```

```python
# 그냥 ffill() 메서드를 쓰는 것도 가능
df.ffill()
```

```python
"""
mean() 메서드는 각 column별 평균 값을 리턴함
fillna()의 argument로 df.mean()을 준 경우
각 column의 평균 값으로 해당 column의 NaN을 대체함
"""
print(df.mean())
df.fillna(df.mean())
```

```python
import pandas_datareader.data as pdr
# 2001/1/2부터 2016/12/30까지 엔화 환율 데이터에 해당

start_date = '2001/1/2'
end_date = '2016/12/30'
fx_jpusdata = pdr.DataReader('DEXJPUS', 'fred', start_date, end_date)
fx_jpusdata.head()
```

```python
# 2016년 4월의 데이터만 확인
fx_jpusdata.loc['2016-04']
```

```python
# 월말의 데이터만 확인
fx_jpusdata.resample('M').last()
```

```python
# 원본 데이터에는 2001/1/6, 2001/1/7의 데이터가 없음
# 이를 생성해야 할 때는 리샘플링 방법을 이용
abcd = fx_jpusdata.resample('D').last()
pd.isnull(abcd).sum()
```

```python
# 2001/1/6, 2001/1/7의 데이터를 직전 날의 값으로 대체
fx_jpusdata.resample('D').ffill().head(10)
```

```python
# 데이터 시프팅
# 날짜별 환율값을 다음 날로 옮김
# 1월 2일의 114.73 -> 1월 3일의 값으로 옮기는 등..
# shift() 메서드의 argument만큼 데이터를 옮김
# shift(2) 라면 2칸씩 옮기는 등..
fx_jpusdata.shift(1).head()
```

```python
# 전날 대비 환율의 비율 구하기
fx_jpusdata_ratio = fx_jpusdata / fx_jpusdata.shift(1)
fx_jpusdata_ratio.head()
```
