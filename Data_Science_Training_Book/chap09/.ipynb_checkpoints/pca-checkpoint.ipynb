```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler
# RandomState 객체를 생성하고, 인자로 1을 넘겨주면서 시드값을 1로 지정
sample = np.random.RandomState(1)

# X는 200 by 2의 행렬
# rand는 0이상 1미만 균일 분포, randn은 표준 정규 분포
X = np.dot(sample.rand(2,2), sample.randn(2, 200)).T

# 표준화
sc = StandardScaler()
X_std = sc.fit_transform(X)

# 두 변수의 상관관계는 0.889, 두 변수는 표준화 되어 평균이 0 분산이 1
print(f"상관 계수: {sp.stats.pearsonr(X_std[:, 0], X_std[:, 1])[0]:.3f}")
plt.scatter(X_std[:, 0], X_std[:, 1])
```

```python
from sklearn.decomposition import PCA

# 변수를 몇개로 줄일 것인지를 n_components의 인자로 넘김
pca = PCA(n_components=2)
# 이 경우 변수를 2개로 줄임

pca.fit(X_std)
```

```python
%precision 3
# components_ 메소드는 PCA를 통해 얻어낸 새로운 변수 공간의 축 방향을 표시
print(pca.components_)
# [-0.707 -0.707]이 제1 주성분 방향, [-0.707 0.707]이 제2 주성분 방향
```

```python
# 각 주성분의 분산을 나타내며, 주성분의 분산의 합은 원래 변수가 가지는 분산의 합과 매우 유사
# 즉, 분산 정보는 유지됨
print(pca.explained_variance_)
pca.explained_variance_.sum()
```

```python
# 각 주성분이 갖는 분산의 비율을 나타내는 메소드
# 0.945 = 1.899 / (1.899 + 0.111)과 같이 계산
# 제1 주성분이 원본 데이터의 정보를 대략 94.5%만큼 유지한다는 의미로 해석
pca.explained_variance_ratio_
```

```python
arrowprops = dict(arrowstyle='->', linewidth=2, shrinkA = 0, shrinkB = 0)

def draw_vector(v0, v1):
    plt.gca().annotate('', v1, v0, arrowprops=arrowprops)

# 원본 데이터 그래프
plt.scatter(X_std[:,0], X_std[:,1], alpha=0.3)

for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)

plt.axis('equal')

# 분산이 최대가 되는 방향이 제1 주성분, 그 다음으로 분산이 커지는 방향이 제2 주성분이 됨
# 벡터 [-0.707 -0.707] 방향이 제1 주성분 방향, 벡터 [-0.707 0.707] 방향이 제1 주성분 방향
# 제1 주성분 벡터와 제2 주성분 벡터는 orthogonal
```

```python
# 주성분 분석 실습
# 유방암 데이터를 사용
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

# 데이터를 malignant(악성)과 benign(양성)으로 구분
# malignant는 cancer.target이 0, benign은 cancer.target이 1
malignant = cancer.data[cancer.target == 0]
benign = cancer.data[cancer.target == 1]
fig, axes = plt.subplots(6, 5, figsize=(20,20))
ax = axes.ravel()

# cancer.data.shape은 (569, 30), cancer.target.shape은 (569,)
# 즉, cancer.data는 569개의 데이터에 대한 30개의 feature의 값
# cancer.data[:, i]는 569개의 데이터에 대한 i번째 feature의 값
for i in range(30):
    _, bins = np.histogram(cancer.data[:,i], bins=50)
    ax[i].hist(malignant[:,i], bins, alpha=0.5)
    ax[i].hist(benign[:,i], bins, alpha=0.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_ylabel('Count')
ax[0].legend(['malignant', 'benign'], loc='best')
fig.tight_layout()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_std = sc.fit_transform(cancer.data)

pca = PCA(n_components=2)
pca.fit(X_std)
# pca.transform은 X_std를 PCA를 통해 축소시킨 차원의 데이터로 변환함
X_pca = pca.transform(X_std)

print(f'X_std shape:{X_std.shape}')
print(f'X_pca shape:{X_pca.shape}')
print(f'Explained variance ratio:{pca.explained_variance_ratio_}')

# pca.explained_variance_ratio_의 합이 1이 아닌 이유는 PCA의 components 수를 2로 두어
# 데이터를 가장 잘 설명하는 변수 2가지만 남겼기 때문
# components의 수가 증가할수록 pca.explained_variance_ratio_의 합도 1에 가까워지고
# components의 수가 30이 되면 pca.explained_variance_ratio_의 합도 1이 됨
```

```python
# 비교
pca = PCA(n_components=30)
pca.fit(X_std)
print(pca.explained_variance_, end = '\n\n')
print(pca.explained_variance_.sum(), end = '\n\n')
print(pca.explained_variance_ratio_, end = '\n\n')
print(pca.explained_variance_ratio_.sum())
# 원본 데이터의 변수가 30개이므로 표준화된 변수의 분산의 합도 30
```

```python
X_pca = pd.DataFrame(X_pca, columns=['pc1','pc2'])
X_pca = pd.concat([X_pca, pd.DataFrame(cancer.target, columns=['target'])], axis=1)
pca_malignant = X_pca[X_pca['target'] == 0]
pca_benign = X_pca[X_pca['target'] == 1]
X_pca
```

```python
ax = pca_malignant.plot.scatter(x='pc1', y='pc2', color='red', label='malignant')
pca_benign.plot.scatter(x='pc1', y='pc2', color='blue', label='benign', ax=ax)

# 그림에서 2개의 주성분으로 목표 변수의 범주를 거의 분류할 수 있다는 것을 알 수 있음
```

```python
# 연습 문제 9-2
from sklearn.datasets import load_iris

iris = load_iris() # iris 품종은 setosa, vesicolor, virginica 3가지
sc = StandardScaler()
iris_std = sc.fit_transform(iris.data)
pca = PCA(n_components=2)
pca.fit(iris_std)
iris_pca = pca.transform(iris_std)
iris_pca = pd.DataFrame(iris_pca, columns=['pc1', 'pc2'])
iris_pca = pd.concat([iris_pca, pd.DataFrame(iris.target, columns=['target'])], axis = 1)
pca_setosa = iris_pca[iris_pca['target'] == 0]
pca_vesicolor = iris_pca[iris_pca['target'] == 1]
pca_virginica = iris_pca[iris_pca['target'] == 2]
ax = pca_setosa.plot.scatter(x = 'pc1', y = 'pc2', color='skyblue', label='setosa')
pca_vesicolor.plot.scatter(x = 'pc1', y = 'pc2', color = 'yellow', label = 'vesicolor', ax = ax)
pca_virginica.plot.scatter(x = 'pc1', y = 'pc2', color = 'grey', label = 'virginica', ax = ax)
```

```python
# iris data의 목표 변수와 4개의 설명 변수 간의 관계 시각화
fig, axes = plt.subplots(2, 2,figsize=(20, 7))
iris_0 = iris.data[iris.target == 0]
iris_1 = iris.data[iris.target == 1]
iris_2 = iris.data[iris.target == 2]

ax = axes.ravel()
for i in range(4):
    _, bins = np.histogram(iris.data[:, i], bins=50)
    ax[i].hist(iris_0[:, i], bins=bins, alpha=0.5)
    ax[i].hist(iris_1[:, i], bins=bins, alpha=0.5)
    ax[i].hist(iris_2[:, i], bins=bins, alpha=0.5)
    ax[i].set_title(iris.feature_names[i])
    ax[i].set_yticks(())
    
ax[0].set_ylabel('Count')
ax[0].legend(['setosa', 'versicolor', 'virginica'], loc = 'best')
fig.tight_layout()
```

```python
for a, b in zip(pca.explained_variance_, pca.components_):
    print(a, b)
```
