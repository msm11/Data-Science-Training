```python
X_data = np.arange(5)
Y_data = np.arange(5, 10)
condition_array = np.array([True, False, False, True, False])
print(X_data)
print(Y_data)
```

```python
# np.where : 조건 배열이 True인 경우 2번째 인자(X_data)에서 값 추출, 조건 배열이 False인 경우 3번째 인자(Y_data)에서 값 추출
print(np.where(condition_array, X_data, Y_data))
```

```python
# 조건 배열에 브로드 캐스팅이 적용 ㅁㅁ
print(np.where(True, X_data, Y_data))
print(np.where(False, X_data, Y_data))
```

```python
# X, Y 데이터에 브로드 캐스팅이 적용
print(np.where(condition_array, X_data, Y_data[1]))
print(np.where(condition_array, X_data[2], Y_data))
```

```python
# 조건 배열의 shape은 (2,5), 데이터의 shape은 (1,5)인 경우로, 브로드 캐스팅 적용 ((3,5)와 (2,5) 사이에서는 연산 x)
condition_array = np.array([[True, False, False, True, False], [True, True, False, False, True]])
print(np.where(condition_array, X_data, Y_data))
```

```python
from scipy import interpolate
x = np.linspace(0, 10, num = 11)
y = np.cos(-x**2/5.0)
# f1은 선형 보간법, f3는 스플라인 3차 보간법에 해당
f1 = interpolate.interp1d(x, y, kind='linear')
f2 = interpolate.interp1d(x, y, kind='cubic')
print(f1(4.5), f2(4.5))

plt.figure(figsize=(15,5))
plt.plot(x,y,'o')
xnew = np.linspace(0, 10, num = 30)
plt.plot(xnew, f1(xnew),'-')
plt.plot(xnew, f2(xnew),'--')
plt.grid(True)
plt.legend(['data', 'linear', 'spline'], loc='lower left')
```

```python
# 연습 문제 5-10
x = np.linspace(0, 10, num=11)
y = np.sin(x**2/5.0)
f = interpolate.interp1d(x, y, kind = 'linear')
plt.plot(x,f(x),'-')
plt.grid(True)
```

```python
# 연습 문제 5-11, 연습 문제 5-12
f2 = interpolate.interp1d(x, y, kind='quadratic')
f3 = interpolate.interp1d(x, y, kind='cubic')
xnew = np.linspace(0, 10, num = 30)
plt.figure(figsize=(15,5))
plt.plot(x,y, 'o')
plt.plot(xnew, f(xnew), '-')
plt.plot(xnew, f2(xnew), '--')
plt.plot(xnew, f3(xnew), ':')
plt.grid(True)
plt.legend(['data', 'linear', 'quadratic', 'cubic'], loc='best')
plt.show()
```

```python
import math
from scipy import integrate
def calcPi(x):
    return 4/(1+x**2)

sp.integrate.quad(calcPi, 0, 1) # quad: 수치 적분 계산, 두 번째 숫자는 오차의 상한값
```

```python
def I(n):
    return sp.integrate.dblquad(lambda t, x: np.exp(-t*x)/t**n, 0, np.inf, lambda x: 1, lambda x: np.inf) # dblquad: 다변수 수치 적분 계산

print(f"{I(1)}\n{I(2)}\n{I(3)}\n{I(4)}")
```
