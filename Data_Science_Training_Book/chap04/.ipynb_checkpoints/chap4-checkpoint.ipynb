```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
# 연습 문제 4-1
np.random.seed(0)
coin_data = np.array([0, 1])
cnt = 1000
coin_flip_data = np.random.choice(coin_data, cnt)
p0 = len(coin_flip_data[coin_flip_data == 0]) / cnt
p1 = len(coin_flip_data[coin_flip_data == 1]) / cnt
print(f'앞면이 나올 확률 = {p0}')
print(f'뒷면이 나올 확률 = {p1}')
```

```python
# 연습 문제 4-2
np.random.seed(0)
a = np.arange(1, 1001)
cntA = 0
for i in range(10000):
    if (np.random.choice(a, 1) >= 1 and np.random.choice(a, 1) <= 100):
        cntA += 1
pA = cntA / 10000

b = np.arange(1, 1000)
cntB = 0
for i in range(10000):
    if (np.random.choice(b, 1) >= 1 and np.random.choice(b, 1) <= 99):
        cntB += 1
pB = cntB / 10000
p = pA*pB
print(f"확률은 {p}")
```

```python
# 연습 문제 4-3
# 양성반응(A)을 보였을 때 질병 X(X)에 걸릴 확률
pA = 0.999*0.03 + 0.001*0.99
pA_and_pX = 0.001*0.99
p = (pA_and_pX) / (pA)
print(f"확률은 {p}")
```

```python
np.random.seed(0)
dice_data = np.arange(1,7)
calc_steps = 1000
dice_rolls = np.random.choice(dice_data, calc_steps)
probability = np.empty(shape = 0)

for i in range(1, 7):
    p = len(dice_rolls[dice_rolls == i]) / calc_steps
    probability = np.append(probability, p)

plt.bar(dice_data, probability)
plt.grid(True)
```

```python
# np.random.seed(0)
# 0 : head(앞) , 1 : tail(뒤)
calc_steps = 8
prob_be_data = np.array([])
coin_data = np.array([0, 1])
coin_flip = np.random.choice(coin_data, calc_steps)

for i in range(2):
    p = len(coin_flip[coin_flip == i]) / calc_steps
    prob_be_data = np.append(prob_be_data, p)

plt.bar(coin_data, prob_be_data, align = 'edge')
plt.xticks(coin_data, ['head','tail'])
plt.grid(True)
```

```python
import scipy.stats as st
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

x, y = np.mgrid[10:100:2, 10:100:2]

pos = np.empty(x.shape + (2, ))
pos[:,:, 0] = x
pos[:,:, 1] = y
rv = multivariate_normal([50,50], [[100,0],[0, 100]])
z = rv.pdf(pos)
fig = plt.figure(dpi = 100)
ax = Axes3D(fig)
ax.plot_wireframe(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.ticklabel_format(style='scientific', axis ='z', scilimits= (0,0))
```

```python
# 주사위를 1000회 던지는 과정을 4회 반복

calc_times = 1000
sample_array = np.array([1,2,3,4,5,6])
number_cnt = np.arange(1, calc_times + 1)

for i in range(4):
    p = np.random.choice(sample_array, calc_times).cumsum() # 시행에 따른 누적합을 구함
    plt.plot(p / number_cnt) # (누적합 / 시행 횟수) == 평균이므로 y축은 평균,  x축은 시행 횟수에 해당하는 그래프를 그림
    plt.grid(True)

# 4가지 그래프 모두 시행 횟수가 커질수록 3.5에 근접해짐
```

```python
np.random.seed(0)
def function_central_theory(N):
    dice_data = np.array([1, 2, 3, 4, 5, 6])
    mean_data = []
    for i in range(1000):
        mean = np.random.choice(dice_data, N).mean() # 표본의 크기가 N인 표본을 추출하여 평균을 구함
        mean_data.append(mean) # mean_data에 표본 평균들을 넣고 그 리스트를 이용하여 히스토그램을 출력
    plt.hist(mean_data)
    plt.grid(True)
        
function_central_theory(3)
```

```python
# 카이 제곱 분포
np.random.seed(0)
for df,c in zip([2, 10, 60], 'bgr'):
    x = np.random.chisquare(df, 1000)  # 자유도가 df인 카이제곱 분포의 원소 1000개를 뽑음
    plt.hist(x, 20, color = c) # 자유도2 : blue / 자유도 10 : green / 자유도 60 : red
    plt.grid(True)
```
